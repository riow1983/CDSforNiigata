# APIを利用したメタデータの取得例 ：				

#### (ア)組織に関するメタデータの取得： 					
- 登録されている組織の一覧 					
　　https://www.data.go.jp/data/api/action/organization_list					
					
- 登録されている組織の一覧(各組織のメタ情報も取得） 					
　　https://www.data.go.jp/data/api/action/organization_list?all_fields=true					
					
- 登録されている組織の一覧（各組織のメタ情報を取得し、且つ組織のデータセット登録数順でソート） 					
　　https://www.data.go.jp/data/api/action/organization_list?all_fields=true&order_by=packages					
					
- 指定した組織「内閣府」の詳細情報 					
　　https://www.data.go.jp/data/api/action/organization_show?id=org_0400					
					
#### (イ)グループに関するメタデータの取得： 					
- 登録されているグループの一覧 					
　　https://www.data.go.jp/data/api/action/organization_list					
					
- 指定したグループ「行財政」の詳細情報					
　　https://www.data.go.jp/data/api/action/group_show?id=gr_1300					
					
#### (ウ)タグに関するメタデータの取得： 					
- タグの一覧 					
　　https://www.data.go.jp/data/api/action/tag_list					
					
- 指定したタグ「統計」を含むデータセット名の一覧（10件取得指定） 					
　　https://www.data.go.jp/data/api/action/package_search?q=*:*&facet.limit=10&facet.field=["title_string"]&fq=tags:統計&rows=0					
					
- キーワード「統計」を含むタグの検索 					
　　https://www.data.go.jp/data/api/action/tag_search?query=統計					
					
#### (エ)データセットやリソースに関するメタデータの取得： 					
- 検索条件に合うデータセット「平成26年_新規登録、更新情報」が属しているグループ名の一覧 					
　　https://www.data.go.jp/data/api/action/package_search?q=*:*&facet.limit=10&facet.field=["groups"]&fq=tags:統計&rows=0					
					
- 検索条件に合うデータセット「高度情報通信ネットワーク社会の形成に関する予算_平成26年度」とそのデータセットに含まれるリソースの一覧 					
　　https://www.data.go.jp/data/api/action/package_search?q=name:cas_20140901_0000					
					
- 指定したデータセット「高度情報通信ネットワーク社会の形成に関する予算_平成26年度」のメタデータ情報 					
　　https://www.data.go.jp/data/api/action/package_show?id=高度情報通信ネットワーク社会の形成に関する予算_平成26年度					
					
- 指定したリソース「高度情報通信ネットワーク社会の形成に関する平成26年度予算について（府省庁別）_HTML」のメタデータ情報 					
　　https://www.data.go.jp/data/api/action/resource_show?id=高度情報通信ネットワーク社会の形成に関する平成26年度予算について（府省庁別）_HTML					
					
- キーワード「統計」を含むデータセットの検索（10件取得指定） 					
　　https://www.data.go.jp/data/api/action/package_search?q=統計&limit=10					
					
- 「name」属性にキーワード「統計」を含むリソースの検索（10件取得指定） 					
　　https://www.data.go.jp/data/api/action/resource_search?query=name:統計&limit=10

# Example

```python
url = 'https://www.data.go.jp/data/api/action/resource_search?query=name:統計&limit=10'
resp = requests.get(url)
assert resp.status_code == 200
```

```python
j = resp.json()
df = pd.DataFrame.from_dict(j["result"]["results"])
```

```python
dls = df["url"][0]
resp = requests.get(dls)
assert resp.status_code == 200
```

```python
with io.BytesIO(resp.content) as fh:
    df = pd.io.excel.read_excel(fh, header=5, sheet_name=0)
```

```python
df.drop(["Unnamed: 0", "Unnamed: 1", "Unnamed: 2"], axis=1, inplace=True)
```

```python
df.rename(columns={"Unnamed: 3":"Index"}, inplace=True)
df.set_index("Index", inplace=True)
```
